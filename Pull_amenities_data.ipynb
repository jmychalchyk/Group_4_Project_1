```python
import pandas as pd
import csv

fl = "data/train.csv"
outfile = "data/amenities.csv"
df = pd.read_csv(fl)
df.head()
#drop bed_type, thumbnail, cleaning fee, name
#has thumbnail to TF

df2col = df[["id",'amenities']]
df2col.set_index('id')
df2col.head()
```

```python

#Creates a unique list of amenities
lst = []
for x in df2col.index:
    dfclean = str(df2col['amenities'][x].replace('{',"").replace('}',"").replace('"','')).split(',')
    for each in dfclean:
        lst.append(each)
dscols = pd.Series(lst).unique()
```

```python
dscols
```

```python
#Creates an empty dataframe to fill
index = df2col['id']
columns = dscols
emptydf = pd.DataFrame(index=index, columns=columns)
emptydf = emptydf.fillna(0) # with 0s rather than NaNs
emptydf.head(5)

```

```python

combined = pd.merge(emptydf,df2col, on='id')
combined.head()
```

```python
templst=[] 
for idx in combined.index:
    clean2 = str(combined['amenities'][idx].replace('{',"").replace('}',"").replace('"','')).split(',') # fix the messed up data
    del templst[:] #Empty the list and repopulate   
    for amenity in clean2:
        templst.append(amenity)
    for each in templst: #'Wireless Internet'
        for colno in range(132): #loop length of columns
            if each == combined.columns[colno]: #If 'Wireless Internet' is the name of the column, put 1
                #print(f'each = {each} : colname = {combined.columns[colno]}: colno = {colno}  : idx = {idx}')
                combined.iloc[idx,colno] = 1
    #break
```

```python
combined.to_csv(outfile)
```
