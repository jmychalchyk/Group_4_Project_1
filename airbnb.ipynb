```python
%matplotlib inline
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# Reference the file where the CSV is located
airbnb_csv_path = "train.csv"

# Import the data into a Pandas DataFrame
airbnb_df = pd.read_csv(airbnb_csv_path)
airbnb_df
```

```python
#get a count by property type
airbnb_df.groupby("property_type")["id"].count()
```

```python
#taking the log price and making it into a easy to read value in terms of $
real_price = []
for x in airbnb_df["log_price"]:
    price = np.exp(x)
    real_price.append(price)
price_series = pd.Series(real_price)
```

```python
airbnb_df["price"] = price_series
```

```python
piechart_data = airbnb_df.groupby("property_type")["id"].count().reset_index()
piechart_data
```

```python
piechart_data = piechart_data.sort_values(by="id",ascending = False).reset_index(drop = True)
piechart_data
```

```python
top_five_data = piechart_data.loc[piechart_data.index < 5]
top_five_data
```

```python

# Creating pie chart based on above data
cities = top_five_data["property_type"]
review_count_by_city = top_five_data["id"]
colors = ["lightskyblue","tan","lightgreen","lightcoral","gold"]
explode = [0,0,0,0,0]
plt.pie(review_count_by_city, explode=explode, labels=cities,colors=colors,autopct="%1.1f%%",shadow=False,startangle=130)
plt.title("% of airbnb properties by property type")
plt.axis("equal")
plt.show()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(top_five_data))
tick_locations = [value+0.4 for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(6,8))
plt.bar(x_axis, top_five_data["id"], color='r', alpha=0.5, align="edge")
plt.xticks(tick_locations, top_five_data["property_type"], rotation="horizontal")

# Set x and y limits
plt.xlim(-0.25, len(x_axis))
plt.ylim(0, max(top_five_data["id"])+1000)

# Set a Title and labels
plt.title("Number of airbnb properties rented by property type")
plt.xlabel("Property Type")
plt.ylabel("Number Rented")
plt.savefig("airbnb_rented_by_property_type")
```

```python
barchart_avgprice = airbnb_df.groupby("property_type")["price"].mean(). reset_index()
barchart_avgprice.sort_values(by="price",ascending = False)
```

```python
top_five_price_avg = pd.merge(top_five_data,barchart_avgprice, on = "property_type", how = "inner")
top_five_price_avg
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(top_five_price_avg))
tick_locations = [value+0.4 for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(10,6))
plt.bar(x_axis, top_five_price_avg["price"], color='g', alpha=0.5, align="edge")
plt.xticks(tick_locations, top_five_price_avg["property_type"], rotation="horizontal")

# Set x and y limits
plt.xlim(-0.25, len(x_axis))
plt.ylim(0, max(top_five_price_avg["price"])+10)
# Set a Title and labels
plt.title("Average Price by Property Type")
plt.xlabel("Property Type")
plt.ylabel("Average Price")
plt.savefig("airbnb_avg_price")
```
