```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = "data/combined_thin.csv"
df = pd.read_csv(file)
df_trim = df.loc[df['review_scores_rating']>0]
```

```python
df_trim.columns
```

```python
df_trim = df_trim[['log_price', 'property_type','number_of_reviews', 'review_scores_rating',
                    'cancellation_policy',
                    'cleaning_fee', 'host_since', 'first_review','last_review',
                    'host_has_profile_pic', 'host_identity_verified', 'host_response_rate',
                    'instant_bookable', 'Accessible', 'Air conditioning', 'Baby friendly',
                    'Beachfront', 'Breakfast','Doorman', 'Elevator',
                    'Family/kid friendly', 'Firm mattress',
                    'Free parking',  'Garden or backyard',
                    'Gym', 'Heating', 'Hot tub', 'Internet', 
                    'Kitchen', 'Laptop friendly workspace',
                    'Lock on bedroom door', 'Long term stays allowed', 'Pool',
                    'Private bathroom', 'Private entrance', 'Private living room',
                    'Smoking allowed',  'TV']]
```

```python
prop_rate        = df_trim[['property_type','review_scores_rating']]
avg_prop_rate    = prop_rate.groupby('property_type').mean()
```

```python
avg_prop_rate
```

```python
prop_rev         = df_trim[['property_type','number_of_reviews']]
sum_prop_rev     = prop_rev.groupby('property_type').sum()
```

```python
sum_prop_rev.join(avg_prop_rate)
```

```python
lst = []
lst = ['Accessible', 'Air conditioning', 'Baby friendly',
                    'Beachfront', 'Breakfast','Doorman', 'Elevator',
                    'Family/kid friendly', 'Firm mattress',
                    'Free parking',  'Garden or backyard',
                    'Gym', 'Heating', 'Hot tub', 'Internet', 
                    'Kitchen', 'Laptop friendly workspace',
                    'Lock on bedroom door', 'Long term stays allowed', 'Pool',
                    'Private bathroom', 'Private entrance', 'Private living room',
                    'Smoking allowed',  'TV']
df_final = sum_prop_rev.join(avg_prop_rate)
for amnty in lst:
    amnty2 = amnty.replace('/',"_").replace(' ',"_")
    df_amnty2=df_trim[['property_type',amnty]].groupby('property_type').sum()
    df_final = df_final.join(df_amnty2.rename(columns={amnty:amnty2}))
```

```python
df_final
```

```python
df_final.to_csv("data/chart_data.csv")
```
